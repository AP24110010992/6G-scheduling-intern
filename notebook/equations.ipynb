```python
import math
```

```python
# 1. Defender Utility
def Ud(D, C):
    alpha = 2
    beta = 1
    return (alpha * D) - (beta * C)
print("Ud =", Ud(10, 3))
```

```python
# 2. Attacker Utility
def Ua(C, D):
    gamma = 3
    delta = 1
    return (gamma * C) - (delta * D)
print("Ua =", Ua(5, 2))
```

```python
# 3. ML Risk Prediction
def R_ml(x):
    return 0.6 * x + 0.4
print("R_ml =", R_ml(0.8))
```

```python
# 4. Risk Blending
def R_blend(Ri, Rml):
    lamda = 0.7
    return (lamda * Ri) + ((1 - lamda) * Rml)
risk = R_blend(0.5, R_ml(0.8))
print("R_blend =", risk)
```

```python
# 5. Detection Probability
def P_detect(Ri, F):
    theta1 = 1.2
    theta2 = 0.8
    exponent = -(theta1 * Ri - theta2 * F)
    return 1 / (1 + math.exp(exponent))
print("P_detect =", P_detect(risk, 0.5))
```
